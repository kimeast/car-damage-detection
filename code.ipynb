# 자동차 파손 영역 검사 및 가격 산출 AI 만들기

```python
import os
import cv2
import json
import torch
import matplotlib.pyplot as plt

from PIL import Image
from src.Models import Unet
```

## 모델 로드 (4가지의 손상 종류에 따라 다 따로 만들었음)

- `[DAMAGE][Breakage_3]Unet.pt` : 파손
- `[DAMAGE][Crushed_2]Unet.pt` : 찌그러짐
- `[DAMAGE][Scratch_0]Unet.pt` : 스크래치
- `[DAMAGE][Seperated_1]Unet.pt` : 이격

```python
# 시간단축을 위해 미리 pre-trained 되어 있는 모델을 활용했음

# weight_path = 'models/[DAMAGE][Breakage_3]Unet.pt'
# weight_path = 'models/[DAMAGE][Crushed_2]Unet.pt'
weight_path = 'models/[DAMAGE][Scratch_0]Unet.pt'
# weight_path = 'models/[DAMAGE][Seperated_1]Unet.pt'

n_classes = 2
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = Unet(encoder='resnet34', pre_weight='imagenet', num_classes=n_classes).to(device)
model.model.load_state_dict(torch.load(weight_path, map_location=torch.device(device)))
model.eval()

print('Loaded pretrained model!')
```

### 전체 이미지 로드 (주의 : 샘플 아닌 raw data로 하면 오래 걸림 & 에러날 확률 높음)

```python
working_directory = "./samples" #json 파일과 jpg 파일이 다 한꺼번에 있는!

# 파일 리스트 가져오기
file_list = os.listdir(working_directory)

# JSON 파일만 필터링
json_files = [file for file in file_list if file.endswith('.json')]

# 각 JSON 파일과 해당하는 이미지 파일을 처리
for json_file in json_files:
    base_filename = os.path.splitext(json_file)[0]  # 확장자 제외한 파일 이름
    img_path = os.path.join(working_directory, base_filename + '.jpg')  # 이미지 파일 경로

    # JSON 파일 열기 및 데이터 읽기
    with open(os.path.join(working_directory, json_file), 'r') as f:
        label_data = json.load(f)

    # 이미지 파일 읽기
    img = cv2.imread(img_path)

    # 이미지 색상 공간 변환 (BGR -> RGB)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 이미지 크기 조정
    img = cv2.resize(img, (256, 256))
    
#     #이미지 표시
#     plt.figure(figsize=(8, 8))
#     plt.imshow(img)
#     plt.title(f"Label: {label_data}")  # 라벨 정보를 제목으로 표시
#     plt.show()
```

### 원하는 n번째 이미지 로드 (이건 빨리 끝남)

```python
n = 1199 # 예를 들어, 세 번째 파일을 보고 싶다면 n = 2 (0부터 시작하는 인덱싱)

# n이 파일 개수 범위 내에 있는지 확인
if n < len(json_files):
    json_file = json_files[n]
    base_filename = os.path.splitext(json_file)[0]  # 확장자 제외한 파일 이름
    img_path = os.path.join(working_directory, base_filename + '.jpg')  # 이미지 파일 경로
    
    # JSON 파일 열기 및 데이터 읽기
    with open(os.path.join(working_directory, json_file), 'r') as f:
        label_data = json.load(f)
    
    # 이미지 파일 읽기
    img = cv2.imread(img_path)
    
    # 이미지 색상 공간 변환 (BGR -> RGB)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # 이미지 크기 조정
    img = cv2.resize(img, (256, 256))
    
    # 이미지 표시
    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.title(f"Label: {label_data}")  # 라벨 정보를 제목으로 표시
    plt.show()
else:
    print("해당 인덱스의 파일이 존재하지 않습니다.")
```

## 이미지 전처리

```python
# 0~255 -> 0~1 값으로 normalize
img_input = img / 255.

# 파이토치에 넣어주기 위해서 채널을 맨 앞으로 당겼음
img_input = img_input.transpose([2, 0, 1])
img_input = torch.tensor(img_input).float().to(device)

#batch 형태로 만들기
img_input = img_input.unsqueeze(0)

# 최종 input size
img_input.shape
```

## 모델 추론

```python
output = model(img_input)

output.shape
```

## 후처리 및 추론 마스크 시각화

```python
img_output = torch.argmax(output, dim=1).detach().cpu().numpy()
img_output = img_output.transpose([1, 2, 0])

plt.figure(figsize=(8, 8))
plt.imshow(img_output) #스크래치는 노란색으로 표기
```

## 결과 (n을 위에서 지정하지 않으면 마지막 사진을 결과로 출력)

```python
#히트맵으로 시각화 좀 더 편하게 만들었음
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 10))

ax[0].imshow(img)
ax[0].set_title('image')
ax[0].axis('off')

ax[1].imshow(img.astype('uint8'), alpha=0.5)
ax[1].imshow(img_output, cmap='jet', alpha=0.5)
ax[1].set_title('output')
ax[1].axis('off')

fig.set_tight_layout(True)
plt.show() 
```

# 여러 형태의 파손 영역 감지

```python
labels = ['Breakage_3', 'Crushed_2', 'Scratch_0', 'Seperated_1']
models = []

n_classes = 2
device = 'cuda' if torch.cuda.is_available() else 'cpu'

for label in labels:
    model_path = f'models/[DAMAGE][{label}]Unet.pt'

    model = Unet(encoder='resnet34', pre_weight='imagenet', num_classes=n_classes).to(device)
    model.model.load_state_dict(torch.load(model_path, map_location=torch.device(device)))
    model.eval()

    models.append(model)

print('Loaded pretrained models!')
```

```python
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (256, 256))

plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
img_input = img / 255.
img_input = img_input.transpose([2, 0, 1])
img_input = torch.tensor(img_input).float().to(device)
img_input = img_input.unsqueeze(0)

fig, ax = plt.subplots(1, 5, figsize=(24, 10))

ax[0].imshow(img)
ax[0].axis('off')

outputs = []

for i, model in enumerate(models):
    output = model(img_input)

    img_output = torch.argmax(output, dim=1).detach().cpu().numpy()
    img_output = img_output.transpose([1, 2, 0])

    outputs.append(img_output)

    ax[i+1].set_title(labels[i])
    ax[i+1].imshow(img_output, cmap='jet')
    ax[i+1].axis('off')

fig.set_tight_layout(True)
plt.show()
```

##  파손 영역 크기 계산

```python
for i, label in enumerate(labels):
    print(f'{label}: {outputs[i].sum()}')
```

## 파손 영역 크기에 따른 가격 산출 (예시)

1픽셀당 가격을 책정하여 영역이 넓을 수록 수리비용이 비싸진다

```python
price_table = [
    100, # Breakage_3
    200, # Crushed_2
    50,  # Scratch_0
    120, # Seperated_1
]

total = 0

for i, price in enumerate(price_table):
    area = outputs[i].sum()
    total += area * price

    print(f'{labels[i]}:\t영역: {area}\t가격:{area * price}원')

print(f'고객님, 총 수리비는 {total}원 입니다!')
```
